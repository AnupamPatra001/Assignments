**Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.**

NLTK and spaCy are both popular Python libraries used for Natural Language Processing (NLP), but they are designed with different purposes in mind.

NLTK (Natural Language Toolkit) is mainly designed for education and research. It provides a wide range of tools such as tokenization, stemming, lemmatization, POS tagging, parsing, and access to many linguistic datasets like WordNet and movie reviews. It gives detailed control over individual NLP tasks, which makes it very useful for understanding the theoretical concepts behind NLP. However, many tasks in NLTK require separate function calls and more manual coding.

On the other hand, spaCy is designed for real-world and industrial applications. It provides built-in, pre-trained models that can perform tokenization, POS tagging, named entity recognition (NER), and dependency parsing in a single processing pipeline. It is more streamlined and developer-friendly compared to NLTK. The API is clean and requires less code to perform complex tasks.

In terms of ease of use, NLTK is helpful for beginners who want to learn how NLP works step by step. spaCy is easier for developers who want to build applications quickly because it offers an integrated pipeline system.

When it comes to performance, spaCy is significantly faster than NLTK. It is optimized using Cython and designed for handling large volumes of text efficiently. NLTK, being mostly pure Python, is slower and not ideal for large-scale production systems.

In conclusion, NLTK is better suited for academic learning and research purposes, while spaCy is more suitable for high-performance, real-world NLP applications.

**Question 2: What is TextBlob and how does it simplify common NLP tasks like
sentiment analysis and translation?**

TextBlob is a simple and beginner-friendly Python library used for Natural Language Processing (NLP). It is built on top of NLTK and Pattern, but it hides much of the complexity and provides an easy-to-use interface. Because of this, it is especially useful for students and beginners who want to perform basic NLP tasks without writing too much code.

TextBlob simplifies common NLP tasks like tokenization, part-of-speech tagging, noun phrase extraction, sentiment analysis, and translation by providing ready-made functions. Instead of manually calling multiple modules and setting up pipelines, users can perform these tasks with just a few lines of code.

For example, in sentiment analysis, TextBlob automatically calculates two values:

Polarity (ranges from -1 to +1), which shows whether the text is negative, neutral, or positive.

Subjectivity (ranges from 0 to 1), which indicates whether the text is more factual or opinion-based.

This means you do not need to train your own model for basic sentiment detection — TextBlob handles it internally using pre-trained models.

Similarly, for translation, TextBlob provides a simple translate() method. You can pass the target language code, and it translates the text automatically. This removes the need to manually integrate external APIs or complex language processing tools.

In summary, TextBlob simplifies NLP by:

Providing a clean and easy API

Reducing the amount of code required

Offering built-in support for common NLP tasks

Making sentiment analysis and translation accessible even to beginners

Therefore, TextBlob is ideal for small projects, quick prototyping, and learning basic NLP concepts.

**Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).**

A Recurrent Neural Network (RNN) is a type of artificial neural network specially designed to handle sequential data such as text, speech, or time-series data. Unlike a normal feedforward neural network, an RNN has the ability to remember previous information in the sequence. This “memory” is what makes it powerful for language-related tasks.

Architecture of RNN:

The basic architecture of an RNN consists of:

Input layer (xₜ) – Takes input at each time step (for example, one word at a time in a sentence).

Hidden layer (hₜ) – This is the core part. It processes the current input along with the previous hidden state.

Output layer (yₜ) – Produces the output for that time step.

The key feature is that the hidden layer has a connection to itself from the previous time step. This creates a loop in the network, which allows information to persist.

Mathematically, the hidden state is calculated as:

hₜ = f(Wxₜ + Uhₜ₋₁ + b)

Here:

xₜ = current input

hₜ₋₁ = previous hidden state

W and U = weight matrices

b = bias

f = activation function (like tanh or ReLU)

This equation shows that the current output depends not only on the current input but also on previous information.

Functioning of RNN:

Let’s understand it practically.

Suppose we have a sentence:
“I am learning NLP”

The RNN reads one word at a time:

At “I” → it processes and stores some information.

At “am” → it uses information from “I” plus current input.

At “learning” → it uses memory from previous words.

At “NLP” → it has context from the entire sequence.

This sequential processing helps in tasks like:

Language modeling

Machine translation

Speech recognition

Sentiment analysis

Important Concept: Backpropagation Through Time (BPTT)

During training, RNN uses a special technique called Backpropagation Through Time. The network is “unrolled” across time steps, and errors are propagated backward to update weights.

Limitation:

Basic RNNs suffer from the vanishing gradient problem, which makes it difficult to learn long-term dependencies. That is why improved versions like LSTM and GRU were introduced.

**Question 5: What is the key difference between LSTM and GRU networks in NLP
applications?**

So, both LSTM and GRU are improved versions of RNN. They were created because normal RNNs forget information when sequences become long. For example, in a long sentence, the meaning of a word at the end may depend on something mentioned much earlier. Basic RNNs struggle with that.

Now the real difference between LSTM and GRU is how they manage memory.

Think of LSTM as a more detailed and careful system. It has three gates — input, forget, and output — and also a separate memory line called the cell state. This separate memory acts like a conveyor belt that carries important information across time steps. Because of this design, LSTM can remember things for a very long time. But since it has more parts and calculations, it is heavier and slower to train.

Now GRU is like a simplified version of LSTM. It combines some of those mechanisms together. Instead of three gates, it has only two — update and reset. It also does not keep a separate cell state. Everything is handled in one hidden state. Because of this simpler structure, GRU trains faster and needs fewer parameters.

In practice, something interesting happens:
Most of the time, GRU performs almost as well as LSTM, especially in tasks like sentiment analysis, text classification, or chat systems. So many developers prefer GRU when they want speed and efficiency. But if the task involves very long sequences or complex language dependencies, LSTM might perform slightly better.

So in simple words:

LSTM is more complex and powerful for long-term memory.
GRU is simpler, faster, and more efficient — but still very strong.



```python
# Question 6: Write a Python program using TextBlob to perform sentiment analysis on
# the following paragraph of text:
# “I had a great experience using the new mobile banking app. The interface is intuitive,
# and customer support was quick to resolve my issue. However, the app did crash once
# during a transaction, which was frustrating"
# Your program should print out the polarity and subjectivity scores.
# (Include your Python code and output in the code box below.)
from textblob import TextBlob

text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

blob = TextBlob(text)

polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print("Polarity:", polarity)
print("Subjectivity:", subjectivity)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

tokens = word_tokenize(text.lower())

words = [word for word in tokens if word.isalpha()]

freq_dist = FreqDist(words)

print("Tokens:")
print(words)

print("\nFrequency Distribution:")
print(freq_dist)
```

```python
# Question 8: Implement a basic LSTM model in Keras for a text classification task using
# the following dummy dataset. Your model should classify sentences as either positive
# (1) or negative (0).
# # Dataset
# texts = [
# “I love this project”, #Positive
# “This is an amazing experience”, #Positive
# “I hate waiting in line”, #Negative
# “This is the worst service”, #Negative
# “Absolutely fantastic!” #Positive
# ]
# labels = [1, 1, 0, 0, 1]
# Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
# this data. You may use Keras with TensorFlow backend.
# (Include your Python code and output in the code box below.)
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
"I love this project",
"This is an amazing experience",
"I hate waiting in line",
"This is the worst service",
"Absolutely fantastic!"
]

labels = [1, 1, 0, 0, 1]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

X = np.array(padded_sequences)
y = np.array(labels)

vocab_size = len(tokenizer.word_index) + 1

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=16, input_length=max_length))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, y, epochs=50, verbose=1)

loss, accuracy = model.evaluate(X, y, verbose=0)
print("Accuracy:", accuracy)

predictions = model.predict(X)
print("Predictions:", predictions)
```

```python
# Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization,
# lemmatization, and entity recognition. Use the following paragraph as your dataset:
# “Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
# development of India’s atomic energy program. He was the founding director of the Tata
# Institute of Fundamental Research (TIFR) and was instrumental in establishing the
# Atomic Energy Commission of India.”
# Write a Python program that processes this text using spaCy, then prints tokens, their
# lemmas, and any named entities found.
# (Include your Python code and output in the code box below.)
import spacy

nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

doc = nlp(text)

print("Tokens and Lemmas:\n")
for token in doc:
    print(f"{token.text:<15} {token.lemma_}")

print("\nNamed Entities:\n")
for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")
```

```python
# Question 10: You are working on a chatbot for a mental health platform. Explain how
# you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
# NLP to understand and respond to user input effectively. Detail your architecture, data
# preprocessing pipeline, and any ethical considerations.
# (Include your Python code and output in the code box below.)

import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

nlp = spacy.load("en_core_web_sm")

texts = [
    "I feel very sad and lonely",
    "I am extremely happy today",
    "I feel anxious about my future",
    "Life is beautiful and exciting"
]

labels = [1, 0, 1, 0]

processed_texts = []

for text in texts:
    doc = nlp(text.lower())
    lemmas = [token.lemma_ for token in doc if token.is_alpha]
    processed_texts.append(" ".join(lemmas))

tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)
sequences = tokenizer.texts_to_sequences(processed_texts)

max_length = max(len(seq) for seq in sequences)
X = pad_sequences(sequences, maxlen=max_length, padding='post')
y = np.array(labels)

vocab_size = len(tokenizer.word_index) + 1

model = Sequential()
model.add(Embedding(vocab_size, 16, input_length=max_length))
model.add(GRU(16))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, y, epochs=30, verbose=1)

loss, accuracy = model.evaluate(X, y, verbose=0)
print("Accuracy:", accuracy)
```
