**Question 1: What are Large Language Models (LLMs) and how do they function?**

Large Language Models (LLMs) are advanced AI systems designed to understand and generate human-like language. Examples include models like ChatGPT and GPT-4. They are called “large” because they are trained on massive amounts of text data—books, articles, websites, and more.

How they function (in simple terms)

Think of an LLM like a highly trained reader that has seen millions of conversations and texts. Instead of memorizing exact answers, it learns patterns in language.

Here's how it works step by step:

Training phase
The model is trained on huge datasets. During this phase, it learns:

Grammar and sentence structure
Meaning of words in different contexts
Relationships between ideas

Technically, it uses a deep learning architecture called a transformer, which helps it understand context better than older models.

Understanding input
When you type a question, the model:
Breaks it into smaller units called tokens (words or parts of words)
Analyzes the context and intent behind your input

Generating output
The model doesn't “think” like humans. Instead, it:

Predicts the most likely next word based on probabilities
Repeats this process word by word to form a complete response

For example, if you type "The sky is…", it predicts words like “blue” based on learned patterns.

Context awareness
Modern LLMs can remember the flow of a conversation (within limits), allowing them to give more relevant and coherent answers.

**Question 5: Compare and contrast Langchain and LamaIndex. What unique problems
does each solve?**

Basic Idea (in simple words)
LangChain -> Helps you build the logic and workflow of an AI application
LlamaIndex -> Helps you connect your data to the AI model

Think of it like this:

LangChain = the brain + workflow manager
LlamaIndex = the data connector + retriever
Key Differences
1. Core Purpose
LangChain:
Focuses on building complete applications using LLMs
(chatbots, agents, pipelines, multi-step reasoning)
LlamaIndex:
Focuses on organizing and retrieving data for LLMs
(documents, PDFs, databases)
2. Problem They Solve
LangChain solves:
How to chain multiple steps together?
How to build agents that take decisions?
How to integrate APIs, tools, memory, and logic?
LlamaIndex solves:
How to use your own data with LLMs?
How to search large document collections efficiently?
How to give context to the model (RAG - Retrieval Augmented Generation)?
3. Working Style
LangChain:
Uses chains (step-by-step pipelines)
Supports agents that can decide actions dynamically
Integrates tools like APIs, databases, and search
LlamaIndex:
Creates indexes from data (like a smart search system)
Retrieves relevant chunks of data when needed
Feeds that data to the LLM for better answers
4. Use Cases
LangChain:
AI chatbots with memory
Autonomous agents
Multi-step workflows (e.g., search -> summarize -> respond)
LlamaIndex:
Document Q&A systems
Knowledge base chatbots
Searching through PDFs, notes, or company data
5. Strength
LangChain: Strong in application logic and orchestration
LlamaIndex: Strong in data handling and retrieval
How They Work Together

In real-world systems, they are often combined:

LlamaIndex -> fetches relevant data
LangChain -> processes it and manages the workflow
Example:
A chatbot for your notes

LlamaIndex retrieves the correct notes
LangChain decides how to answer and interact

```python
# Question 6: Implement a basic Langchain pipeline using OpenAI’s LLM to answer
# questions based on a user input prompt.
# (Include your Python code and output in the code box below.)
# Install (run once)
!pip install -U langchain langchain-community transformers sentence-transformers

# Imports
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# -----------------------------
# Step 1: Initialize Local LLM
# -----------------------------
pipe = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80
)

llm = HuggingFacePipeline(pipeline=pipe)

# -----------------------------
# Step 2: Create Prompt Template
# -----------------------------
prompt = PromptTemplate.from_template(
    "Answer the following question clearly:\n\nQuestion: {question}\nAnswer:"
)

# -----------------------------
# Step 3: Create LangChain Pipeline
# -----------------------------
chain = prompt | llm | StrOutputParser()

# -----------------------------
# Step 4: Run Example
# -----------------------------
question = "What is Artificial Intelligence?"
response = chain.invoke({"question": question})

print("Q:", question)
print("A:", response)
```

```python
# Install required libraries
!pip install requests langchain langchain-community transformers torch

# Imports
import requests
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate

# Step 1: Create a function to call Weather API
def get_weather(city):
    url = f"https://wttr.in/{city}?format=3"   # Free weather API (no key needed)
    response = requests.get(url)
    return response.text

# Step 2: Load free LLM (no API key)
pipe = pipeline("text-generation", model="gpt2")
llm = HuggingFacePipeline(pipeline=pipe)

# Step 3: Prompt template
prompt = PromptTemplate.from_template(
    "Given the weather information: {weather}\n"
    "Answer the user question: {question}\n"
    "Provide a clear and helpful response."
)

# Step 4: Create chain
chain = prompt | llm

# Step 5: User input
city = "Bhubaneswar"
question = "Should I carry an umbrella today?"

# Step 6: Fetch weather data
weather_data = get_weather(city)

# Step 7: Run LLM
response = chain.invoke({
    "weather": weather_data,
    "question": question
})

# Step 8: Output
print("Weather Data:", weather_data)
print("\nLLM Response:\n", response)
```

```python
# Question 8: Create a LamaIndex implementation that indexes a local text file and
# retrieves answers from it.
# (Include your Python code and output in the code box below.)
# Install (run once)
!pip install -U llama-index llama-index-embeddings-huggingface \
               transformers sentence-transformers

# Imports
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from transformers import pipeline

# -----------------------------
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# -----------------------------
# Step 2: Create local text file
# -----------------------------
with open("info.txt", "w") as f:
    f.write("""
    Data Science involves collecting, analyzing, and interpreting data.
    It uses techniques from statistics, machine learning, and data analysis.
    Python is widely used in Data Science due to its simplicity and libraries.
    """)

# -----------------------------
# Step 3: Load and index file
# -----------------------------
documents = SimpleDirectoryReader(input_files=["info.txt"]).load_data()
index = VectorStoreIndex.from_documents(documents)

# -----------------------------
# Step 4: Retrieval (NO LLM inside)
# -----------------------------
retriever = index.as_retriever(similarity_top_k=1)

# -----------------------------
# Step 5: Simple answer generation
# -----------------------------
generator = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80
)

def ask_question(question):
    # Retrieve relevant text
    nodes = retriever.retrieve(question)
    context = nodes[0].text[:300]

    # Create prompt
    prompt = f"""
    Answer based on the context.

    Context: {context}
    Question: {question}
    Answer:
    """

    result = generator(prompt)[0]["generated_text"]
    return result.split("Answer:")[-1].strip()

# -----------------------------
# Run Example
# -----------------------------
q = "Why is Python used in Data Science?"
print("Q:", q)
print("A:", ask_question(q))
```

```python
# Question 9: Demonstrate combining Langchain with LamaIndex to create a simple
# document-based Q&A chatbot.
# (Include your Python code and output in the code box below.)
# Install required libraries
# Install (run once)
!pip install -U llama-index llama-index-embeddings-huggingface \
               langchain transformers sentence-transformers

# Imports
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from transformers import pipeline
from langchain_core.prompts import PromptTemplate

# -----------------------------
# Step 1: LlamaIndex (Retrieval)
# -----------------------------

# Local embedding (no OpenAI)
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Create sample document
with open("data.txt", "w") as f:
    f.write("""
    Artificial Intelligence (AI) is transforming industries by automating tasks.
    Machine Learning is a subset of AI that enables systems to learn from data.
    Deep Learning uses neural networks for complex problem solving.
    """)

# Load and index document
documents = SimpleDirectoryReader(input_files=["data.txt"]).load_data()
index = VectorStoreIndex.from_documents(documents)

# Retriever (safe, no LLM here)
retriever = index.as_retriever(similarity_top_k=1)

# -----------------------------
# Step 2: LangChain (Generation)
# -----------------------------

# Local LLM
generator = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80
)

# Prompt template (LangChain)
prompt = PromptTemplate.from_template(
    "Answer based on the context.\n\nContext: {context}\nQuestion: {question}\nAnswer:"
)

# -----------------------------
# Step 3: Combine both
# -----------------------------

def chatbot(question):
    # Get relevant context from LlamaIndex
    nodes = retriever.retrieve(question)
    context = nodes[0].text[:300]

    # Create prompt
    final_prompt = prompt.format(context=context, question=question)

    # Generate answer using LLM
    output = generator(final_prompt)[0]["generated_text"]

    return output.split("Answer:")[-1].strip()

# -----------------------------
# Run Example
# -----------------------------
q = "What is Machine Learning?"
print("Q:", q)
print("A:", chatbot(q))
```

```python
# Install
!pip install -U llama-index llama-index-embeddings-huggingface \
               transformers sentence-transformers

# Imports
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from transformers import pipeline

# ✅ Only embeddings (NO LLM inside LlamaIndex)
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Create document
with open("legal.txt", "w") as f:
    f.write("""
    This contract establishes that the supplier is responsible for delivery delays.
    The liability clause states that damages will be compensated under certain conditions.
    Both parties must adhere to the terms stated above.
    """)

# Index
documents = SimpleDirectoryReader(input_files=["legal.txt"]).load_data()
index = VectorStoreIndex.from_documents(documents)

# 🔥 PURE retrieval (no LLM → no crash)
retriever = index.as_retriever(similarity_top_k=1)

# ✅ Separate LLM (safe)
generator = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80
)

# Chat function
def chatbot(question):
    nodes = retriever.retrieve(question)
    context = nodes[0].text[:300]   # limit context

    prompt = f"""
    Answer the question based only on the context.

    Context: {context}
    Question: {question}
    Answer:
    """

    result = generator(prompt)[0]["generated_text"]
    return result.split("Answer:")[-1].strip()

# Run
q = "What does the liability clause state?"
print("Q:", q)
print("A:", chatbot(q))
```
