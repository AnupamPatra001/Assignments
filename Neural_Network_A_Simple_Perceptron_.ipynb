**Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**

Deep Learning is a branch of machine learning that tries to teach computers to learn in a way similar to how humans learn. It uses layered structures called neural networks, where each layer learns something more detailed from the data. Because of these many layers, it is called deep learning. Deep learning is commonly used in things like face recognition, voice assistants, language translation, and self-driving cars.

How Deep Learning Evolved (in Simple Terms)

Early days: Scientists first created very simple brain-like models that could learn basic patterns, but they were limited and couldn't solve complex problems.

Improvement phase: Later, better training methods were discovered, allowing networks to learn more than simple patterns. Still, computers were slow and data was limited.

Modern era: With powerful computers, large datasets, and better algorithms, deep learning became practical and highly successful. This is why it is widely used today.

How Deep Learning Is Different from Traditional Machine Learning

In traditional machine learning, humans usually have to tell the computer what features to look at (for example, edges or colors in an image).

In deep learning, the model automatically learns these features on its own from raw data.

Traditional machine learning works well with small datasets, while deep learning performs best with large amounts of data.

Deep learning models are generally more accurate for complex tasks but require more computation.

**Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?**

A Perceptron is the most basic model of an artificial neuron. You can think of it as a very simple decision-maker that looks at some inputs and then decides yes or no, true or false, or 0 or 1.

Basic Architecture (How It Is Built)

A perceptron has four main parts:

Inputs – the values it receives (for example, marks, height, pixels, etc.).

Weights – numbers that show how important each input is.

Bias – an extra value that helps adjust the final decision.

Activation function – usually a step function that gives the final output.

The perceptron first multiplies each input by its weight, adds them all together, adds the bias, and then checks whether the result crosses a certain limit.

How a Perceptron Works (In Easy Steps)

Take the input values.

Multiply each input by its weight.

Add all the results and include the bias.

Pass the sum through an activation function.

If the value is high enough, output 1; otherwise, output 0.

During learning, the perceptron compares its output with the correct answer and slightly changes the weights if it makes a mistake. Over time, it learns to make better decisions.

Limitations of a Perceptron

It can only solve simple problems where data can be separated by a straight line.

It cannot solve complex problems like the XOR problem.

It has only one layer, so it cannot learn complicated patterns.

It works only for two output classes (yes/no type problems).

**Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.**

An activation function helps a neural network decide how strongly a neuron should respond to the information it receives. Without activation functions, a neural network would just perform simple straight-line (linear) calculations and would not be able to learn complex patterns.

In short, activation functions:

Add non-linearity to the network

Help the model learn complex relationships

Decide whether a neuron should “fire” or not

Sigmoid, Tanh, and ReLU – Easy Comparison
1. Sigmoid Function

Think of Sigmoid as a smooth on-off switch.

Output is always between 0 and 1

Often used when the output needs to represent probability

Pros:

Easy to understand

Good for yes/no type outputs

Cons:

Learning becomes slow for deep networks

Gradients become very small (vanishing gradient problem)

2. Tanh Function

Tanh is similar to sigmoid but more balanced.

Output ranges from –1 to +1

Zero-centered, which helps learning

Pros:

Better than sigmoid for hidden layers

Faster learning than sigmoid

Cons:

Still suffers from vanishing gradients

Slower than ReLU

3. ReLU (Rectified Linear Unit)

ReLU works like a simple rule: if the input is positive, pass it; otherwise, output zero.

Output is 0 for negative values, otherwise equals the input

Pros:

Very simple and fast

Works well for deep networks

Reduces vanishing gradient problem

Cons:

Some neurons may stop learning if they always get negative input (dead neurons)

In Short

Sigmoid → good for probability outputs, but slow learning

Tanh → better than sigmoid, but still has limitations

ReLU → most popular today because it is fast and effective

**Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.**

Both loss function and cost function are used to measure how wrong a neural network is. The difference is mainly scope.

Loss Function (Error for One Example)

A loss function tells us how wrong the model is for one single data point.

Think of it like:

“How much mistake did I make on this one question?”

Example:

Suppose a model predicts a student’s marks as 70

Actual marks are 80

The loss function measures the error for this single student

Common examples:

Mean Squared Error (MSE)

Binary Cross Entropy

Cost Function (Error for All Examples)

A cost function is the overall error of the model across the entire dataset.
It is usually the average of all individual losses.

Think of it like:

“On average, how badly am I performing in the whole exam?”

Example:

Calculate loss for every student

Take the average of all those losses

That average value is the cost

Key Difference in One Line

Loss function → error for one data point

Cost function → error for all data points together

Simple Analogy (Easy to Remember)

Loss = marks lost in one question

Cost = total marks lost in the entire paper

**Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.**

An optimizer is the method a neural network uses to improve itself during training. Its main job is to adjust the model’s weights and bias so that the error (loss) becomes as small as possible.

In simple words, an optimizer answers this question:

“How should the model change its weights to make better predictions next time?”

Optimizers decide:

How much to change the weights (learning rate)

In which direction to change them

Comparison of Gradient Descent, RMSprop, and Adam
1. Gradient Descent

This is the basic optimizer.

It updates weights by moving in the direction that reduces error.

Uses the same learning rate for all parameters.

Pros:

Simple and easy to understand

Works well for small and simple problems

Cons:

Can be slow

May get stuck in local minima

Sensitive to learning rate choice

2. RMSprop

RMSprop improves Gradient Descent by adjusting the learning rate.

It changes the learning rate based on recent gradients.

Helps when gradients change rapidly.

Pros:

Faster convergence than Gradient Descent

Works well for non-stationary problems

Cons:

Needs careful tuning

Slightly more complex

3. Adam (Adaptive Moment Estimation)

Adam combines ideas from Momentum and RMSprop.

It keeps track of past gradients and squared gradients.

Automatically adjusts learning rates.

Pros:

Very fast and efficient

Works well for deep and large networks

Little tuning required

Cons:

Uses more memory

Sometimes not the best for final fine-tuning

```python
# Question 6: Write a Python program to implement a single-layer perceptron from
# scratch using NumPy to solve the logical AND gate.
# (Include your Python code and output in the code box below.)
import numpy as np

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.1

def step_function(x):
    return 1 if x >= 0 else 0

epochs = 10
for _ in range(epochs):
    for i in range(len(X)):
        output = np.dot(X[i], weights) + bias
        prediction = step_function(output)
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

print("AND Gate Results:")
for i in range(len(X)):
    result = step_function(np.dot(X[i], weights) + bias)
    print("Input:", X[i], "Output:", result)
```

```python
# Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
# using Matplotlib.
# (Include your Python code and output in the code box below.)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.figure()
plt.plot(x, sigmoid)
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("Output")
plt.show()

plt.figure()
plt.plot(x, relu)
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("Output")
plt.show()

plt.figure()
plt.plot(x, tanh)
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("Output")
plt.show()
```

```python
# Question 8: Use Keras to build and train a simple multilayer neural network on the
# MNIST digits dataset. Print the training accuracy.
# (Include your Python code and output in the code box below.)
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5, batch_size=32)

train_loss, train_accuracy = model.evaluate(X_train, y_train)
print("Training Accuracy:", train_accuracy)
```

```python
# Question 9: Visualize the loss and accuracy curves for a neural network model trained
# on the Fashion MNIST dataset. Interpret the training behavior.
# (Include your Python code and output in the code box below.)
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_data=(X_test, y_test)
)

plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.show()

plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training and Validation Accuracy")
plt.show()
```
