**Question 1: What is hypothesis testing in statistics?**

Hypothesis testing in statistics is a formal process to make inferences or decisions regarding a population parameter using sample information.

It enables us to decide if the sample evidence is sufficient to either accept or reject an assumption (hypothesis) about the population.

Key Points:

Hypothesis - An assertion regarding a population parameter.

Null hypothesis (H₀): The default assumption (e.g., "there is no effect" or "no difference").

Alternative hypothesis (H₁ or Ha): The statement we are seeking to test (e.g., "there is an effect" or "a difference exists").

Test Statistic - A sample data calculated numerical value (such as z, t, χ², F) employed to test the hypothesis.

Significance Level (α): The probability threshold (typically 0.05) for determining whether or not to reject H₀.

p-value: Probability of seeing the sample data (or something more extreme) when H₀ is true.

If p ≤ α, reject H₀.

If p > α, fail to reject H₀.

Decision: Based on the evidence, we reject the null hypothesis or fail to reject it (we never "accept" it completely).

Example:

A company states that the average life of its battery is 100 hours.

H₀: μ = 100 (no difference)

H₁: μ ≠ 100 (different from 100)
If a sample has mean = 95 hours with statistical support (p < 0.05), we reject H₀ and conclude the battery life is not 100 hours.

**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

The null hypothesis (H₀) is like the “default belief” — it assumes nothing special is happening.

Example: A coin is fair, a medicine has no effect, or the average score is still 50.

The alternative hypothesis (H₁ or Ha) is the challenger — it says something different is happening.

Example: The coin is biased, the medicine works, or the average score is not 50.

Think of it like a courtroom trial:

H₀ (Null Hypothesis): The accused is innocent until proven guilty.

H₁ (Alternative Hypothesis): The accused is guilty.

The job of hypothesis testing is like the judge and jury — they look at the evidence (data). If the evidence is strong enough, they reject the null (innocent) and accept the alternative (guilty). If the evidence is weak, they stick with the null.

**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

The significance level (α) in hypothesis testing is basically the cutoff point we set before making a decision. It answers the question: “How much chance of being wrong am I willing to accept?”

Think of it like this:
If you’re a teacher checking whether a student really did their own homework, you might say, “I’ll only accuse them of cheating if I’m at least 95% sure.”
That means you’re accepting a 5% risk (α = 0.05) of being wrong.

In statistics:

If the evidence from the data (the p-value) is smaller than α, you say, “The evidence is strong enough, I reject the null hypothesis.”

If the p-value is larger than α, you say, “The evidence isn’t strong enough, I’ll stick with the null hypothesis for now.”

So, the significance level is like your rule for doubt. A lower α (like 0.01) means you’re being very strict and want stronger evidence before rejecting the null. A higher α (like 0.10) means you’re more relaxed and okay with weaker evidence.

**Question 4: What are Type I and Type II errors? Give examples of each.**

In hypothesis testing, there are two kinds of mistakes we can make when making a decision based on data:

Type I Error (False Alarm)
This happens when we reject the null hypothesis even though it’s actually true.
In other words: we think “something is happening,” but in reality, nothing has changed.
Example: A fire alarm goes off, but there’s no fire. You panicked for no reason.
Statistical example: A medical test says a healthy person has a disease.

Type II Error (Missed Detection)
This happens when we fail to reject the null hypothesis even though it’s actually false.
In other words: we think “nothing is happening,” but in reality, there is something.
Example: There’s actually a fire, but the alarm doesn’t ring. You miss the danger.
Statistical example: A medical test says a sick person is healthy.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

Z-test:-

A Z-test is used when:

You already know the population standard deviation (σ).

You have a large sample size (generally n > 30).

Think of it like this: if you have a big enough group and you know a lot about the population, the Z-test works well. It uses the normal distribution (the bell curve).

Example: A factory claims their bulbs last 1,000 hours with a known standard deviation of 100 hours. If you take a large sample of 50 bulbs, you’d use a Z-test to check if the claim holds.

T-test:-

A T-test is used when:

The population standard deviation is unknown.

You have a small sample size (n ≤ 30).

Here, we don't know as much, so we rely on the sample data itself to estimate variability. That's why the T-test uses the t-distribution, which has "fatter tails" to allow for more uncertainty.

Example: You are testing a new medicine on only 15 patients. You don't know the population standard deviation, so you would use a T-test to check if the medicine had a real effect.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# (Include your Python code and output in the code box below.)

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)
mu_0 = 50

z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", round(sample_mean, 3))
print("Sample Std Dev:", round(sample_std, 3))
print("Sample Size:", n)
print("Z-statistic:", round(z_stat, 3))
print("p-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# (Include your Python code and output in the code box below.)
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)

data = np.random.normal(loc=50, scale=5, size=100)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

confidence = 0.95
alpha = 1 - confidence
z_critical = stats.norm.ppf(1 - alpha/2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Sample Mean:", round(sample_mean, 3))
print("95% Confidence Interval: ({:.3f}, {:.3f})".format(lower_bound, upper_bound))

plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Lower 95% CI')
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper 95% CI')
plt.title("Normal Distribution with 95% CI")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
# (Include your Python code and output in the code box below.)
import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z = (data - mean) / std
    return z

# Example dataset
data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5])

# Calculate Z-scores
z = z_scores(data)
print("Z-scores:", np.round(z, 3))

# Plot histogram
plt.hist(z, bins=8, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (0)')
plt.legend()
plt.show()
```
